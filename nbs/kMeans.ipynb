```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [6, 6]
```

```python
X, y = datasets.make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=1.1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', lw=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', lw=1)
```

```python
k = 3
```

```python
from mlscratch.k_means import kMeans
```

```python
mdl = kMeans(k=3)
```

```python
y_preds = mdl.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_preds, edgecolors='k', lw=1)
```

```python
from sklearn.cluster import KMeans
```

```python
sk_mdl = KMeans(n_clusters=3)
```

```python
sk_preds = sk_mdl.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=sk_preds, edgecolors='k', lw=1)
```
