```python
from mlscratch.imports import *
from mlscratch.linear_algebra import LinearRegression
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

%pylab inline

pylab.rcParams['figure.figsize'] = [16.0, 6.0]
plt.style.use("ggplot")
```

```python
# read data from text file
filename = "../data/nba.txt"
data = np.genfromtxt(filename, dtype=None, delimiter=',',
                     names=True, encoding=None)
x = data['x']
y = data['y']
```

```python
slope, intercept, error, x, y, y_model = LinearRegression.least_squares(x, y)
```

```python
plt.scatter(x, y)
plot([min(x), max(x)], [intercept + slope*min(x), intercept + slope*max(x)],
     color='k', linestyle='-', linewidth=2)

plt.xlabel("PM");
plt.ylabel("Team Winning Percentage");
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print("statsmodels OLS results: ", results.params)
print("LSLF results: ", intercept, slope)
```
