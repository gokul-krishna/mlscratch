```python
from mlscratch.imports import *
from mlscratch.linear_algebra import SVD_Regression, LinearRegression
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

%pylab inline

pylab.rcParams['figure.figsize'] = [16.0, 6.0]
plt.style.use("ggplot")
cmap = plt.cm.coolwarm
```

```python
# read data from text file
filename = "../data/nba.txt"
data = np.genfromtxt(filename, dtype=None, delimiter=',',
                     names=True, encoding=None)
x = data['x']
y = data['y']
```

```python
slope_LS, intercept_LS, error_LS, _, _, y_model_LS = LinearRegression.least_squares(x, y)
slope_SVD, intercept_SVD, y_model_SVD = SVD_Regression.SVD_regression(x, y)

plt.scatter(x,y)
plt.plot(x, y_model_SVD, color = "b", lw=7, alpha=0.5)
plt.plot(x, y_model_LS, color = "r")
custom_lines = [Line2D([0], [0], color=cmap(0.), lw=5),
                Line2D([0], [0], color=cmap(1.), lw=1)]
plt.legend(custom_lines, ['SVD Appx.', 'Least Squares'])
```
